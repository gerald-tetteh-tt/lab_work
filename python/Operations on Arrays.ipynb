# Question 1

```python
import numpy as np
```

```python
data_2016 = np.genfromtxt("AAPL-2016.csv",delimiter=",")
data_2017 = np.genfromtxt("AAPL-2017.csv",delimiter=",")
data_2016 = data_2016[1:,1:]
data_2017 = data_2017[2:,1:]
print(data_2016.shape)
print(data_2017.shape)
```

```python
price_volume_2016 = data_2016[:,[4,6]]
price_volume_2017 = data_2017[:,[4,6]]
```

```python
# axis of 0 sums down the columns
print(price_volume_2016.sum(axis=0))
print(price_volume_2017.sum(axis=0))
```

# Question 2

```python
change = price_volume_2017 - price_volume_2016
print(change.sum(axis=0))
```

# Question 3

```python
print(change)
np.set_printoptions(formatter={"float": "{: 6.2f}".format})
print(change)
```

# Question 4

```python
combined_years = np.concatenate((data_2016,data_2017))
print(combined_years)
```
