```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_2016 = pd.read_csv("AAPL-2016.csv")
data_2017 = pd.read_csv("AAPL-2017.csv")
data_2017 = data_2017.drop([0])
print(data_2017.shape)
print(data_2016.shape)
print(data_2016.head())
print(data_2017.head())
```

```python
plot_data_2016 = data_2016.loc[:39]
#print(plot_data_2016)
```

```python
x_data = np.arange(1,41)
plt.plot(x_data,plot_data_2016["Volume"])
```

```python
open_price_2017 = data_2017["Open"]
open_price_2016 = data_2016["Open"]
```

```python
x_data = np.arange(1,251)

plt.title("AAPL 2017 vs 2016")

plt.plot(x_data,open_price_2017,color="red",linestyle=":",label="2017")
plt.plot(x_data,open_price_2016,color="blue",linestyle=":",label="2016")

plt.legend()
```
